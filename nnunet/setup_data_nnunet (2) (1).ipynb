```python
import os
from PIL import Image
import numpy as np

# Dossiers d'entrée
base_dir = r"\\ps-home.insa-lyon.fr\users\home\selkhalifi\Documents\PIR\nnunet_data\nnUNet_raw\Dataset300_Sep"
images_dir = os.path.join(base_dir, "imagesTr")
labels_dir = os.path.join(base_dir, "labelsTr")

def convert_to_uint8(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".png"):
            path = os.path.join(directory, filename)
            try:
                # Charger et convertir en niveaux de gris (1 canal)
                img = Image.open(path).convert("L")
                arr = np.array(img)

                # Conversion en uint8 si nécessaire
                arr = np.clip(arr, 0, 255).astype(np.uint8)

                # Sauvegarde en remplacement
                img_out = Image.fromarray(arr)
                img_out.save(path)

                print(f"Converti en uint8 : {filename}")
            except Exception as e:
                print(f"Erreur sur {filename} : {e}")

# Traitement des images et labels
print("Conversion des IMAGES...")
convert_to_uint8(images_dir)

print("\nConversion des LABELS...")
convert_to_uint8(labels_dir)
```

```python
import os
from PIL import Image
import numpy as np

# === Dossiers contenant déjà les fichiers ===
image_folder = "./nnUNet_raw/Dataset300_Test/imagesTs"
label_folder = "./nnUNet_raw/Dataset300_Test/labelsTs"

# === Renommage des images ===
for filename in os.listdir(image_folder):
    if filename.endswith(".png") and "-Flair" in filename:
        # Exemple : Patient-52_52-Flair.png
        patient_id = filename.replace("-Flair.png", "")
        src_path = os.path.join(image_folder, filename)
        dst_filename = f"{patient_id}_0000.png"
        dst_path = os.path.join(image_folder, dst_filename)
        
        os.rename(src_path, dst_path)
        print(f"[Image renommée] {filename} -> {dst_filename}")

# === Traitement et renommage des labels ===
for filename in os.listdir(label_folder):
    if filename.endswith(".png") and "-LesionSeg-Flair" in filename:
        # Exemple : Patient-52_52-LesionSeg-Flair.png
        patient_id = filename.replace("-LesionSeg-Flair.png", "")
        src_path = os.path.join(label_folder, filename)
        dst_filename = f"{patient_id}.png"
        dst_path = os.path.join(label_folder, dst_filename)

        # Charger et binariser le masque
        label_img = Image.open(src_path).convert("L")
        label_np = np.array(label_img)

        # Convertir 255 -> 1 et forcer à 0/1 (uint8)
        label_bin = (label_np > 0).astype(np.uint8)

        # Sauvegarder
        Image.fromarray(label_bin * 255).save(dst_path)  # pour visualisation, 0/255
        print(f"[Label corrigé + renommé] {filename} -> {dst_filename}")

        # Optionnel : supprimer l'ancien fichier
        os.remove(src_path)
```

```python
import os
import numpy as np
import nibabel as nib
from PIL import Image

def convert_png_to_nii(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    for fname in os.listdir(input_dir):
        if fname.endswith(".png"):
            img_path = os.path.join(input_dir, fname)
            img = Image.open(img_path).convert("L")
            img_np = np.array(img, dtype=np.uint8)
            # Convert 255 -> 1 (binaire)
            img_bin = (img_np > 0).astype(np.uint8)
            # Add singleton dimension: (1, H, W) to mimic slice
            img_bin = img_bin[np.newaxis, :, :]
            # Save as NIfTI
            out_path = os.path.join(output_dir, fname.replace(".png", ".nii.gz"))
            nib.save(nib.Nifti1Image(img_bin, affine=np.eye(4)), out_path)
            print(f"Converti : {fname} -> {os.path.basename(out_path)}")

# Dossiers source et cible
png_labels = "D:/PIR_SEP/nnunet_data/nnUNet_raw/Dataset300_Sep/labelsTs"
png_preds = "D:/PIR_SEP/nnunet_data/nnUNet_raw/Dataset300_Sep/preds"

nii_labels = "D:/PIR_SEP/nnunet_data/nnUNet_raw/Dataset300_Sep/labels_nifti"
nii_preds = "D:/PIR_SEP/nnunet_data/nnUNet_raw/Dataset300_Sep/preds_nifti"

# Conversion des deux dossiers
convert_png_to_nii(png_labels, nii_labels)
convert_png_to_nii(png_preds, nii_preds)
```
