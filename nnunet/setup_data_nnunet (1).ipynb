```python
import os
from PIL import Image
import numpy as np

# Dossiers d'entr√©e
base_dir = r"\\ps-home.insa-lyon.fr\users\home\selkhalifi\Documents\PIR\nnunet_data\nnUNet_raw\Dataset300_Sep"
images_dir = os.path.join(base_dir, "imagesTr")
labels_dir = os.path.join(base_dir, "labelsTr")

def convert_to_uint8(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".png"):
            path = os.path.join(directory, filename)
            try:
                # Charger et convertir en niveaux de gris (1 canal)
                img = Image.open(path).convert("L")
                arr = np.array(img)

                # Conversion en uint8 si n√©cessaire
                arr = np.clip(arr, 0, 255).astype(np.uint8)

                # Sauvegarde en remplacement
                img_out = Image.fromarray(arr)
                img_out.save(path)

                print(f"Converti en uint8 : {filename}")
            except Exception as e:
                print(f"Erreur sur {filename} : {e}")

# Traitement des images et labels
print("Conversion des IMAGES...")
convert_to_uint8(images_dir)

print("\nConversion des LABELS...")
convert_to_uint8(labels_dir)
```

```python
import os
from PIL import Image
import numpy as np

images_dir = r"\\ps-home.insa-lyon.fr\users\home\selkhalifi\Documents\PIR\nnunet_data\nnUNet_raw\Dataset300_Sep\imagesTr"
labels_dir = r"\\ps-home.insa-lyon.fr\users\home\selkhalifi\Documents\PIR\nnunet_data\nnUNet_raw\Dataset300_Sep\labelsTr"

for fname in os.listdir(images_dir):
    if fname.endswith("_0000.png"):
        try:
            img_path = os.path.join(images_dir, fname)
            lbl_path = os.path.join(labels_dir, fname.replace("_0000", ""))
            img = Image.open(img_path).convert("L")
            lbl = Image.open(lbl_path).convert("L")

            img_arr = np.array(img).astype(np.uint8)
            lbl_arr = np.array(lbl).astype(np.uint8)

            assert img_arr.shape == lbl_arr.shape, f"Taille diff√©rente pour {fname}"
            assert img_arr.dtype == np.uint8, f"dtype image invalide : {fname}"
            assert lbl_arr.dtype == np.uint8, f"dtype label invalide : {fname}"
        except Exception as e:
            print(f"Probl√®me d√©tect√© avec {fname}: {e}")
```
